```python
data['ORDER_NUMBER'] = data['ORDER_NUMBER'].astype(str)
data['ITEM_FAMILY'] = data['ITEM_FAMILY'].astype(str)
data['PL'] = data['PL'].astype(str)
data['SEGMENT'] = data['SEGMENT'].astype(str)
data['ITEM_NUMBER'] = data['ITEM_NUMBER'].astype(str)
data['INVOICE_NUMBER'] = data['INVOICE_NUMBER'].astype(str)
data['INVOICE_YEAR'] = data['INVOICE_YEAR'].astype(str)
data['INVOICE_YEAR_MONTH'] = data['INVOICE_YEAR_MONTH'].astype(str)
data['CUSTOMER_NUMBER'] = data['CUSTOMER_NUMBER'].astype(str)
```

```python
data = data.loc[data['INVOICE_YEAR'] == "2021"]
```

**Monetary Value: what is the sum of monetary value from this client**

```python
Monetary_value = data.groupby("CUSTOMER_NAME").sum("CA").reset_index()[['CUSTOMER_NAME', 'CA']]
print(Monetary_value)
```

**Frequency: how many times does the client take this action?**

```python
Frequency = data.groupby('CUSTOMER_NAME').ORDER_NUMBER.nunique().reset_index().rename(columns={'ORDER_NUMBER':'ORDER_COUNT'})
print(Frequency)
```

**Recency: when is the last time the client takes an action**

```python
most_recent_order = data.sort_values('INVOICE_DATE').groupby('CUSTOMER_NAME').tail(1)
```

```python
# add column for now dat 
d_final = date(2021,12,31)
most_recent_order['now_date'] = pd.to_datetime(d_final)
```

```python
most_recent_order['recency'] = most_recent_order['now_date'] - most_recent_order['INVOICE_DATE']
most_recent_order = most_recent_order[['CUSTOMER_NAME', 'recency']]
most_recent_order
```

```python
most_recent_order.dtypes
```

**Merge the 3**

```python
merge_1 = most_recent_order.merge(Frequency)
merge_2 = merge_1.merge(Monetary_value)
```

```python
RFM_dataset = merge_2[merge_2['CA'] >= 0 ]
```

```python
RFM_dataset = RFM_dataset.rename({'ORDER_COUNT' : 'Frequency'}, axis=1)
```

```python
RFM_dataset
```

```python
RFM_dataset["FrequencyScore"] = pd.cut(RFM_dataset["Frequency"], 
                                       bins=[-1, np.percentile(RFM_dataset["Frequency"], 20), 
                                             np.percentile(RFM_dataset["Frequency"], 40), 
                                             np.percentile(RFM_dataset["Frequency"], 60), 
                                             np.percentile(RFM_dataset["Frequency"], 80),
                                             RFM_dataset["Frequency"].max()], labels=[1, 2, 3, 4, 5]).astype("int")
```

```python
RFM_dataset["RecencyScore"] = pd.cut(RFM_dataset["recency"].dt.days, 
                                       bins=[-1, np.percentile(RFM_dataset["recency"].dt.days, 20), 
                                             np.percentile(RFM_dataset["recency"].dt.days, 40)+1, 
                                             np.percentile(RFM_dataset["recency"].dt.days, 60), 
                                             np.percentile(RFM_dataset["recency"].dt.days, 80),
                                             RFM_dataset["recency"].dt.days.max()], labels=[5, 4, 3, 2, 1]).astype("int")
```

```python
RFM_dataset["MVScore"] = pd.cut(RFM_dataset["CA"], 
                                       bins=[-1, np.percentile(RFM_dataset["CA"], 20), 
                                             np.percentile(RFM_dataset["CA"], 40),
                                             np.percentile(RFM_dataset["CA"], 60), 
                                             np.percentile(RFM_dataset["CA"], 80), 
                                             RFM_dataset["CA"].max()], labels=[1, 2, 3, 4, 5]).astype("int")
```

```python
RFM_dataset
```

**Overall RFM score**

```python
RFM_dataset["RFM"] = RFM_dataset["RecencyScore"] + RFM_dataset["FrequencyScore"] + RFM_dataset["MVScore"]
```

```python
RFM_dataset.sort_values("RFM", ascending=False)
```

```python
RFM_dataset.describe()
```

```python
print("Nombre total de client : ", len(np.unique(RFM_dataset['CUSTOMER_NAME'])))
```

## Create segment of customers

**First segment = Champions : Ordered recently, ordered often and spend the most**

```python
champions = RFM_dataset.loc[(RFM_dataset['RFM'] == 15)]
print("There are ", len(champions['CUSTOMER_NAME']), " champions clients")
champions
```

**Loyal clients: ordered frequently and not long ago**

```python
Loyal_clients = RFM_dataset.loc[(RFM_dataset['FrequencyScore'] >=4) & (RFM_dataset['RecencyScore'] >=3) & (RFM_dataset['RecencyScore'] >=3) & (RFM_dataset['RFM']<15)]

print("Loyal clients : ", len(Loyal_clients['CUSTOMER_NAME']))
Loyal_clients
```

**Need attion = Below Average Recency, Frequency**

```python
RFM_dataset['RecencyScore'].mean()
```

```python
Need_attention= RFM_dataset.loc[(RFM_dataset['FrequencyScore'] < RFM_dataset['FrequencyScore'].mean()) & (RFM_dataset['FrequencyScore'] >1) & (RFM_dataset['RecencyScore'] < RFM_dataset['RecencyScore'].mean()) & (RFM_dataset['RecencyScore']>1)]

print("Need Attention clients : ", len(Need_attention['CUSTOMER_NAME']))
Need_attention
```

**Second Segment = At Risk : ordered a long time ago, but used to order frequently**

```python
At_risk = RFM_dataset.loc[(RFM_dataset['RecencyScore'] < 3) & (RFM_dataset['FrequencyScore'] >= 3)]
print("Need Attention clients : ", len(At_risk['CUSTOMER_NAME']))
At_risk
```

**Hibernating: last ordered was a long time ago and low number of orders. May be lost clients**

```python
Hibernating = RFM_dataset.loc[(RFM_dataset['FrequencyScore'] <=2) & (RFM_dataset['RecencyScore'] <2)]

print("Loyal clients : ", len(Hibernating['CUSTOMER_NAME']))
Hibernating
```

**Potential loyalists: Recent customers with average frequency**

```python
Potential_loyalists = RFM_dataset.loc[(RFM_dataset['FrequencyScore'] ==3) & (RFM_dataset['RecencyScore'] >=4)]
print("Potential Loyalists :", len(Potential_loyalists['CUSTOMER_NAME']))
Potential_loyalists
```

**Create column for category of RFM segmentation**

```python
conditions = [
    (RFM_dataset['RFM'] == 15),
    (RFM_dataset['FrequencyScore'] >= 4) & (RFM_dataset['RecencyScore'] >= 3),
    (RFM_dataset['FrequencyScore'] <= 3) & (RFM_dataset['FrequencyScore'] >1) & (RFM_dataset['RecencyScore']<= 3) & (RFM_dataset['RecencyScore']>1),
    (RFM_dataset['FrequencyScore'] >=3) & (RFM_dataset['RecencyScore'] <3),
    (RFM_dataset['FrequencyScore'] <=2) & (RFM_dataset['RecencyScore']<=2),
    (RFM_dataset['FrequencyScore'] ==1) & (RFM_dataset['RecencyScore'] == 3),
    (RFM_dataset['FrequencyScore'] ==3) & (RFM_dataset['RecencyScore'] >=4),
    (RFM_dataset['FrequencyScore']<=2) & (RFM_dataset['RecencyScore']>4) & (RFM_dataset['MVScore']>=4)
    ]
choices = ['Champion', "Loyal Client", "Need Attention", "At risk", "Hibernating", "About to sleep", "Potential Loyalist", "Promising"]
```

```python
RFM_dataset['RFM_Segment'] = np.select(conditions, choices, default='Non affected')
```
